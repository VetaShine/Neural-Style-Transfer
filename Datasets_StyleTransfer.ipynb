# **Datasets for Style Transfer**

Для решения задачи переноса стиля нам потребуется два датасета - первый для стилевых изображений, второй для основных изображений. 

**Dataset for Style Images**

Рассмотрим используемый в этой работе датасет для стилевых изображений - Best Artworks of All Time. Он содержит коллекцию работ 50 самых влиятельных художников всех времён. Самих же работ художников в общем насчитывается порядка 8684 объектов. Загрузить его можно с сайта Kaggle по ссылке: https://www.kaggle.com/datasets/ikarus777/best-artworks-of-all-time.

Этот набор данных содержит три файла:

- **artist.csv**: набор данных информации для каждого художника;

- **images.zip**: коллекция изображений (полного размера), разделённых на папки по художникам и последовательно пронумерованных;

- **resized.zip**: такая же коллекция, но свойства изображений изменены и извлечены из структуры папок.

При создании Style Transfer нам не понадобятся все три файла, а лишь один - images.zip.

**Dataset for Content Images**

Рассмотрим используемый в этой работе датасет для основных изображений - Google Scraped Image Dataset. Он содержит коллекцию разнообразных фотографий. Самих же фотографий в общем насчитывается порядка 35242 объектов. Загрузить его можно с сайта Kaggle по ссылке: https://www.kaggle.com/datasets/duttadebadri/image-classification.

- **images.zip**: коллекция изображений, разделённых на папки по темам: architecture, art and culture, food and drinks, travel and adventure;
